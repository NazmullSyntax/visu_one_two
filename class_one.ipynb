```python
import pandas as pd
import numpy as np

# Create sample dataset
data = {
    "Name": ["Alice", "Bob", np.nan, "David"],
    "Age": [25, 30, 35, np.nan],
    "Salary": [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)


# 1. Check missing values
print("\nMissing Values:")
print(df.isnull().sum())


# 2. Fill missing values
df["Name"] = df["Name"].fillna("Unknown")
df["Age"] = df["Age"].fillna(df["Age"].mean())

print("\nAfter Filling Missing Values:")
print(df)


# 3. Remove duplicate data
df = df.drop_duplicates()

print("\nAfter Removing Duplicates:")
print(df)


# 4. Add new column
df["Salary_In_Lakh"] = df["Salary"] / 100000

print("\nFinal Processed Data:")
print(df)
```

```python
# Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Create sample data
data = {
    "Name": ["Nazmul", "Rahim", "Karim", "Sakib", "Tamim"],
    "Age": [22, 25, np.nan, 23, 24],
    "Salary": [30000, 45000, 50000, np.nan, 40000],
    "Department": ["IT", "HR", "IT", "Finance", "HR"]
}


# Convert dictionary to DataFrame
df = pd.DataFrame(data)


# Show original data
print("Original Data:")
print(df)


# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())


# Fill missing values with mean
df["Age"] = df["Age"].fillna(df["Age"].mean())

df["Salary"] = df["Salary"].fillna(df["Salary"].mean())


# Add new column using NumPy
df["Bonus"] = np.array(df["Salary"]) * 0.10


# Filtering data
high_salary = df[df["Salary"] > 40000]

print("\nHigh Salary Employees:")
print(high_salary)


# Calculate mean salary
average_salary = df["Salary"].mean()

print("\nAverage Salary:")
print(average_salary)


# Data information
print("\nData Information:")
print(df.info())


# Visualization using Seaborn

plt.figure(figsize=(8,5))

sns.barplot(
    x="Name",
    y="Salary",
    data=df
)

plt.title("Employee Salary Chart")
plt.xlabel("Employee Name")
plt.ylabel("Salary")

plt.show()
```
